```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import time
```

```python
# Part 1
# Write a program to extract the color histogram of each of the 2,000 images. 
# Choose the parameters required with justifications. 
# Implement your own histogram code and compare its results with open-source API like OpenCV and numpy.

# # computes histogram for 3 channels separately and concatenates them
def compute_hist(img, bins=256):
    # compute size of one bin
    bin_size = 256/bins
    # convert pixel values to a bin numbers
    img = (img/bin_size).astype(int)
    hist_bgr = []
    for i in range(3):
        # create dict with 0 values for keys in range from 0 to 256 
        channel_hist = dict.fromkeys(list(range(bins)), 0)
        # count unique values for one channel
        unique, counts = np.unique(img[:,:,i], return_counts=True)
        # update dict with new count values
        channel_hist.update(dict(zip(unique, counts)))
        hist_bgr.append(list(channel_hist.values()))
    return np.concatenate(hist_bgr)

# computes histogram using OpenCV
def compute_hist_cv(img, bins=256):
    hist_bgr = []
    for i in range(3):
        hist = cv2.calcHist([img],[i],None,[bins],[0,256])
        hist_bgr.append(hist)
    return np.concatenate(hist_bgr).flatten()

# computes histogram using numpy
def compute_hist_numpy(img, bins=256):
    hist_bgr = []
    for i in range(3):
        hist = np.histogram(img_test[:,:,i], bins=bins, range=(0,256))[0]
        hist_bgr.append(hist)
    return np.concatenate(hist_bgr)

# reads all images from dir and returns list of histograms
def get_all_hists(img_dir, bins=256):
    hist_dict = {}
    for filename in os.listdir(img_dir):
        img = cv2.imread(os.path.join(img_dir, filename))
        hist_dict[filename] = compute_hist(img, bins)
    return hist_dict
        
data_path = '../../Homework1'
hist_dict = get_all_hists(data_path)
```

```python
## Part 2
# Write a program to measure the L2 distance between color histograms of two images.

# returns L2 dist
def compute_L2(hist1, hist2):
    """ Computes and returns L2 distance between two 1D arrays """
    return np.sqrt(((hist1 - hist2)**2).sum())
```

```python
## Part 3
# Use 5 images shown above (ukbench00004.jpg; ukbench00040.jpg; ukbench00060.jpg; ukbench00588.jpg; ukbench01562.jpg) 
# as queries. For each query image, find 10 best matches from the 2,000 images based on the color histogram similarity.
# Plot the query image and the 10 returned matches (use icons of reduced resolution to save space).

# returns filenames of n best matches
def find_best_matches(hist_dict, query_filename, n):
    # get tuples (filename, distance) with L2 distance 
    # from the hist of query_filename image to other histograms in hist_dict
    distances = [(name, compute_L2(hist_dict[query_filename], hist))
                          for name, hist in hist_dict.items() if name != query_filename]
    # sort by distance
    distances.sort(key = lambda x: x[1])
    # return n filenames with minimal coresponding distance
    return [x[0] for x in distances[:n]]

# shows and saves plot of query images and matched images
def plot_matches(query_img, matched_imgs, save_name, show=True):
    fig, axs = plt.subplots(3, 5, figsize = (15,5))
    axs = np.ravel(axs)
    for ax in axs:
        ax.set_xticks([])
        ax.set_yticks([])
        ax.axis('off')
    axs[2].imshow(query_img)
    axs[2].set_title('query image')
    for ind, ax in enumerate(axs[5:]):
        ax.imshow(matched_imgs[ind])
        ax.set_title(f'match {ind+1}')
    plt.savefig(f'./query_results/{save_name}.png')
    plt.show()

query_images_names = ['ukbench00004.jpg', 'ukbench00040.jpg', 'ukbench00060.jpg', 
                     'ukbench00588.jpg', 'ukbench01562.jpg']
best_matches_list = []
for query_name in query_images_names:
    best_matches = find_best_matches(hist_dict, query_name, 10)
    best_matches_list.append(best_matches)
    query_img = cv2.imread(os.path.join(data_path, query_name))
    query_img = cv2.cvtColor(query_img, cv2.COLOR_BGR2RGB)
    matched_imgs = [cv2.imread(os.path.join(data_path, filename)) for filename in best_matches]
    matched_imgs = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB) for img in matched_imgs]
    plot_matches(query_img, matched_imgs, f'best_matches_{query_name[:-4]}')
```

```python
# Part4
# Write a program to measure and plot the P-R curve for each query.

true_matches = [['ukbench00005.jpg', 'ukbench00006.jpg', 'ukbench00007.jpg'],
                ['ukbench00041.jpg', 'ukbench00042.jpg', 'ukbench00043.jpg'],
                ['ukbench00061.jpg', 'ukbench00062.jpg', 'ukbench00063.jpg'],
                ['ukbench00589.jpg', 'ukbench00590.jpg', 'ukbench00591.jpg'],
                ['ukbench01560.jpg', 'ukbench01561.jpg', 'ukbench01563.jpg']]

# Plots Precision-Recall Curve for every query from the previous task with a maximum number of matches equal 10
def plot_PR_curve():
    fig, ax = plt.subplots(2,3, figsize=(15,9))
    plt.subplots_adjust(hspace=0.3, wspace=0.3)
    ax = np.ravel(ax)
    ax[5].remove()
    for ind, img in enumerate(query_images_names):
        P = []
        R = []
        for n in range(1, 11):
            TP = len(set(best_matches_list[ind][:n]) & set(true_matches[ind]))
            FP = n - TP
            FN = 3 - TP
            TN = 2000 - n - FN
            P.append(TP / (TP+FP))
            R.append(TP / (TP+FN))
        ax[ind].plot(P, R)
        ax[ind].set_title(img)
        ax[ind].set_xlabel('precision')
        ax[ind].set_ylabel('recall')
    plt.suptitle('Precision-Recall curves')
        
plot_PR_curve()
```
