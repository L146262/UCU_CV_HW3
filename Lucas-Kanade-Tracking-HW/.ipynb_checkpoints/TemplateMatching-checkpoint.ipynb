```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sum of absolute differences
def SAD(p1, p2):
    return np.abs(p1 - p2).sum()

# sum of squared differences
def SSD(p1, p2):
    return ((p1-p2)**2).sum()

# normalized cross correlation
def NCC(p1, p2):
    return (p1*p2).sum() / np.sqrt(np.sum(p1**2)*np.sum(p2**2))


def match_template(target_img, img, metric_func):
    window_h, window_w = target_img.shape
    img_h, img_w = img.shape
    half_h, half_w = window_h//2, window_w//2
    bordered_img = cv2.copyMakeBorder(img,
                    top=half_h, bottom=half_h,
                    left=half_w, right=half_w,
                    borderType=cv2.BORDER_REPLICATE)
    similarity_map = np.empty((img_h, img_w))
    for i in range(img_h):
        for j in range(img_w):
            similarity_map[i,j] = metric_func(target_img, 
                                      bordered_img[i:i+window_h,
                                                   j:j+window_w])
    pick_ind = np.unravel_index(similarity_map.argmin(), similarity_map.shape)
    return similarity_map, pick_ind

def draw_frame(img, y, x, frame_h, frame_w):
    new_img = img.copy()
    start_point = (x - frame_w//2, y - frame_h//2)
    end_point = (x + frame_w//2, y + frame_h//2)
    new_img = cv2.rectangle(new_img, start_point, end_point, (255, 0, 0), 2) 
    return new_img
```

```python
# match template and
def test_matching(target_img_path, test_imgs_paths, target_coordinates):
    img = cv2.imread(target_img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    test_imgs = [cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2GRAY) 
                 for path in test_imgs_paths]
    
    y1, x1, y2, x2 = target_coordinates

    target_img = img[y1:y2, x1:x2]
    target_plot = draw_frame(img, (y2+y1)//2, (x2+x1)//2,*target_img.shape)
    plt.imshow(target_plot, cmap='gray');
    plt.title('Target');
    
    for metric in [SAD, SSD, NCC]:
        frame_center_coord = [match_template(target_img, im, metric)[1] 
                                    for im in test_imgs]
            
        fig, axs = plt.subplots(2,3, figsize=(15,7))
        axs = axs.flatten()
        for i, ax in enumerate(axs):
            new_img = draw_frame(test_imgs[i], *frame_center_coord[i], *target_img.shape)
            ax.imshow(new_img, cmap='gray')
        plt.suptitle(f'Matching Results using {metric.__name__}')
    
```

```python
img_path = '../../David/img/0306.jpg'
test_imgs_paths = ['../../David/img/0307.jpg', '../../David/img/0309.jpg', '../../David/img/0312.jpg', 
                   '../../David/img/0320.jpg', '../../David/img/0330.jpg', '../../David/img/0340.jpg']

target_coordinates = (50, 90, 130, 155)

test_matching(img_path, test_imgs_paths, target_coordinates)
```
